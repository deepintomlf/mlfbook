## **绘制激活函数及其导数**

```
def sigmoid(x):
    ''' 定义sigmoid函数 '''
    return 1/(1 + np.exp(-x))

def tanh(x):
    ''' 定义tanh函数 '''
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

def ReLU(x):
    ''' 定义ReLU函数 '''
    return np.maximum(0,x)

def gradient(x, f, h):
    ''' 计算梯度 '''
    h=1e-4
    grad = (f(x+h)-f(x-h))/2*h
    return grad
  
def sigmoid_gradient(x):
    ''' 计算梯度 '''
    return sigmoid(x)*(1-sigmoid(x))
  
def tanh_gradient(x):
    ''' 计算梯度 '''
    return (1-tanh(x)*tanh(x))
 
def ReLU_gradient(x):
    ''' 计算梯度 '''
    return (x>0)
```

```
import numpy as np

# 生成模拟数据
x = np.array(np.linspace(-4,4))
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_ReLU = ReLU(x)

# 计算激活函数的梯度
grad_sigmoid = sigmoid_gradient(x)
grad_tanh = tanh_gradient(x)
grad_ReLU = ReLU_gradient(x)
```

```
import matplotlib.pyplot as plt
from google.colab import files

def plot_activation_func(x, y ,grad, str):
  plt.figure(figsize=(18,6))
  plt.subplot(1,2,1)
  plt.plot(x, y, 'b', linewidth=2)
  plt.legend(['f(x)'], loc='upper left', fontsize=24)
  plt.xticks(fontsize=20)
  plt.yticks(fontsize=20)
  plt.subplot(1,2,2)
  if (str == 'Relu function'):
    zero_loc = np.argmin(np.abs(x))+1
    print(zero_loc)
    plt.plot(x[:zero_loc], grad[:zero_loc], 'r', linewidth=2)
    plt.plot(0, 0, 'ro', fillstyle='none', linewidth=1.5)
    plt.plot(x[zero_loc:], grad[zero_loc:], 'r', linewidth=2)
  else:
    plt.plot(x, grad, 'r', linewidth=2)
  
  plt.legend(['df/dx'], loc='upper left', fontsize=24)
  plt.xticks(fontsize=20)
  plt.yticks(fontsize=20)  
  plt.savefig(str, dpi = 1000)
  files.download(str+'.png')
  
```

```
plot_activation_func(x,y_sigmoid, grad_sigmoid,'Sigmoid function')
plot_activation_func(x,y_tanh, grad_tanh, 'Tanh function')
plot_activation_func(x,y_ReLU, grad_ReLU, 'Relu function')
```
