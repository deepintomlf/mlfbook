## **绘制激活函数及其导数**

```
def sigmoid(x):
    ''' 定义sigmoid函数 '''
    return 1/(1 + np.exp(-x))

def tanh(x):
    ''' 定义tanh函数 '''
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

def ReLU(x):
    ''' 定义ReLU函数 '''
    return np.maximum(0,x)

def gradient(x, f):
    ''' 计算梯度 '''
    h=1e-4
    grad = (f(x+h)-f(x-h))/2*h
    return grad
```

```
import numpy as np

# 生成模拟数据
x = np.array(np.linspace(-4,4))
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_ReLU = ReLU(x)

# 计算激活函数的梯度
grad_sigmoid = gradient(x, sigmoid)
grad_tanh = gradient(x, tanh)
grad_ReLU = gradient(x, ReLU)
```

```
import matplotlib.pyplot as plt

def plot_activation_func(x, y ,grad):
  plt.figure(figsize=(18,6))
  plt.subplot(1,2,1)
  plt.plot(x, y, 'b', linewidth=2)
  plt.legend(['f(x)'], loc='upper left', fontsize=24)
  plt.xticks(fontsize=20)
  plt.yticks(fontsize=20)
  plt.subplot(1,2,2)
  plt.plot(x, grad, 'r', linewidth=2)
  plt.legend(['df/dx'], loc='upper left', fontsize=24)
  plt.xticks(fontsize=20)
  plt.yticks(fontsize=20)
```

```
plot_activation_func(x,y_sigmoid, grad_sigmoid)
plot_activation_func(x,y_tanh, grad_tanh)
plot_activation_func(x,y_ReLU, grad_ReLU)
```

```
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.plot(x, y_sigmoid, 'b', linewidth=2)
plt.legend(['f(x)'], loc='upper left', fontsize=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.subplot(1,2,2)
plt.plot(x, grad_tanh, 'r', linewidth=2)
plt.legend(['df/dx'], loc='upper left', fontsize=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig('sigmoid.png', dpi=500)

plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.plot(x, y_tanh, 'b', linewidth=2)
plt.legend(['f(x)'], loc='upper left', fontsize=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.subplot(1,2,2)
plt.plot(x, grad_tanh, 'r', linewidth=2)
plt.legend(['df/dx'], loc='upper left', fontsize=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig('tanh.png', dpi=500)

plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.plot(x, y_ReLU, 'b', linewidth=2)
plt.legend(['f(x)'], loc='upper left', fontsize=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.subplot(1,2,2)
plt.plot(x, grad_ReLU, 'r', linewidth=2)
plt.legend(['df/dx'], loc='upper left', fontsize=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```
